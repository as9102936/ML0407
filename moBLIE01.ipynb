```python
# 發佈時間
a_tag = soup.select("div.l-listTable__td:nth-of-type(2) div.o-fNotes")
for i in a_tag:
    potime = i.text
    print(potime)
```

```python
# 最後回覆時間
a_tag = soup.select("div.l-listTable__td:nth-of-type(3) div.o-fNotes")
for i in a_tag:
    lasttime = i.text
    print(lasttime)
```

```python
# 回應人數
a_tag = soup.select("div.o-fMini")
peo = []
for i in a_tag:
    peo = i.string
    print(peo)
```
