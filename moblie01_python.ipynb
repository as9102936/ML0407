```python
#抓內文
# test 

import requests
import csv
import io
import ssl
import os
import time
from bs4 import BeautifulSoup
import traceback
import pandas as pd
# import osdict
import errno


test_h1 = []
test_time = []
test_href = []
test_article = []
test_reply_num = []

for web_url in allwebpages:

    page = requests.get(web_url,
                    headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36"})
    soup = BeautifulSoup(page.text, 'html.parser')

    a_tag = soup.select("h1")#主題標簽
    a_tag1 = soup.select(".l-publish__content .c-tool__check span, div.l-articlePage:nth-of-type(n+3) .c-tool__check span")#回應人數標簽

    a_tag2 =soup.select(".l-publish__content span.o-fNotes:nth-of-type(1), div.l-articlePage:nth-of-type(n+3) span.o-fNotes:nth-of-type(1)")
    # 發佈時間
    a_tag3 = soup.select("div[itemprop='articleBody'], article.u-gapBottom--max")#內文
    for i in range (len(a_tag1)):
        
        test_href.append(str(web_url))
        test_h1.append(a_tag[0].text)
        test_time.append(a_tag2[i].text)
        test_article.append(a_tag3[i].text.replace("\n", "").replace(" ", "").replace("\xa0",""))
        test_reply_num.append(a_tag1[i].text)

print(len(test_article))
print(len(test_h1))
print(len(test_time))
print(len(test_href))
print(len(test_reply_num))
```
